```python
cols=[var for var in df.columns if df[var].isnull().mean() < 0.05 and df[var].isnull().mean()>0]
```

```python
cols 
```

```python
df[cols].sample(5)
```

```python
len(df[cols].dropna())/len(df)
```

```python
new_df=df[cols].dropna()
df.shape,new_df.shape
```

```python
new_df.hist(bins=50,density=True,figsize=(12,42))
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
#original data
df['training_hours'].hist(bins=50,ax=ax,density=True,color='red')
#data after cca
new_df['training_hours'].hist(bins=50,ax=ax,density=True,color='green',alpha=0.8)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
#original data
df['training_hours'].plot.density(color='red')
#after cca
new_df['training_hours'].plot.density(color='green')
```

# CCA in cartegorical data

```python
df['education_level'].value_counts()
```

```python
df['enrolled_university'].value_counts()
```

cca for enrolled_university

```python
temp=pd.concat([
df['enrolled_university'].value_counts()/len(df),
new_df['enrolled_university'].value_counts()/len(new_df)
],
axis=1)
temp.columns=['original','cca']
temp
```

cca for enducation level

```python
temp=pd.concat([
    df['education_level'].value_counts()/len(df),
    new_df['education_level'].value_counts()/len(df)   
],
axis=1)
temp.columns=['original','cca']
temp
```
